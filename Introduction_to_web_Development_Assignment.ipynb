# **Q1. Explain the main components of client-server architecture and their functions.**
- There are 2 main components of Client-Server architecture which are as follows:

    - **Client**
    - **Server**

  These both are used to provide communication and data exchange between devices. Following is the brief introduction of main components of client-server architecture and their Functions:

## - Introduction to Client
      Client is any device or software application that request services or resource from a server, in computer context.
      Example : Like Web Browser, Email Client, etc.
      Request consists of "Type of Resource" or "service being requested".
      Basically Client sends a request containing the specific details of what it is looking for.
      Client Devices : "Computers", "Laptops", "Smartphones", "Tablets" etc.

## - Introduction to Server
      Servers are devices or sofware application that provide service or resources to clients.
      All the request came from client is being processed by server, it performs necessary actions and Calculations, and then returns the result back to client in requested format.
      Example : like hosting websites, storing and retrieving files, or managing email communication.
      Servers are connected to internet or computer network, allowing them to communicate with clients located anywhere in the world.
      Servers are designed to be highly reliable and available, with advanced features such are redundancy and load balancing to ensure that they can handle large Traffic and requests.

# **Q3. Describe the role of a web browser's developer tools in web development.**
- Developer Tools help developers debug, test, and optimize web applications. These tools provide real-time insights into a webpage's HTML, CSS, JavaScript, network activity, and performance.

- There are developer tools like Elements, Console, Network, Performance, Source, Application, Security.

- Key Features of DevTools:
	1.	Elements Tab : Allows developers to inspect and modify HTML & CSS in real time.
	2.	Console Tab : Displays errors, warnings, and logs related to JavaScript execution.
	3.	Network Tab : Monitors API requests, loading times, and HTTP status codes.
	4.	Performance Tab : Helps analyze and improve the loading speed and responsiveness of web pages.
	5.	Sources Tab : Enables debugging JavaScript code using breakpoints.
	6.	Application Tab : Shows cookies, local storage, session storage, and indexedDB data.
	7.	Security Tab : Provides insights into SSL certificates and potential security vulnerabilities.


# **Q4. What are the advantages of using a Version Control System like Git in a development project?**

- Git is a distributed version control system that helps teams manage code efficiently. It provides several advantages:
  1. Code Versioning
    - Keeps track of changes made to the code.
    - Allows developers to revert to previous versions if needed.
  2. Collaboration & Teamwork
    - Multiple developers can work on the same project without conflicts.
    - Branching and merging allow parallel development.
  3. Backup & Security
    - Code is stored in repositories like GitHub, GitLab, or Bitbucket, preventing data loss.
    - Version control ensures safe rollbacks if something breaks.
  4. Branching & Merging
    - Developers can create feature branches without affecting the main codebase.
    - Changes can be merged after proper testing.
  5. Continuous Integration/Deployment (CI/CD)
    - Automates testing and deployment processes.
    - Ensures smooth updates without breaking existing functionality.

  
# **Q5. Compare and contrast a text editor and an IDE, highlighting their key features and uses.**

- Both text editors and Integrated Development Environments (IDEs) are used for coding, but they serve different purposes.

- Text Editor : A lightweight tool primarily used for writing and editing code.
  - Examples:
      - VS Code
      - Sublime Text
      - Notepad++
  - Key Features:
      - Fast and lightweight.
      - Syntax highlighting and auto-completion.
      - Extensions and plugins for customization.

- Integrated Development Environment (IDE) : A full-featured software suite that provides tools for coding, debugging, and compiling.
  - Examples:
      - Visual Studio
      - JetBrains IntelliJ IDEA
      - Eclipse
  - Key Features:
      - Built-in debugger, compiler, and terminal.
      - Advanced code navigation and refactoring tools.
      - Integrated version control (Git).
